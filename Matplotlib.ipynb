# **Line Graph**

```
import matplotlib.pyplot as plt
import numpy as np
```

```
x = np.arange(10)
y1 = x**2
y2 = 2*x+3

print(x)
print(y1)
print(y2)
```

```
themes = plt.style.available
```

```
len(themes)
```

```
print(themes)
```

```
themes = plt.style.available
for ele in themes:
    plt.style.use(ele)
    plt.plot(x,y1,color ='green',label = 'Quadratic',marker='o' )
    plt.plot(x,y2,color ='red' ,label = 'Linear',linestyle = 'dashed')
    plt.xlabel('Time')
    plt.ylabel('Distance')
    plt.legend(loc='upper left')
    plt.title(ele)
    plt.show()
```

# **Scatter Plot**

```
themes = plt.style.available
for ele in themes:
    plt.figure(figsize=(7,7))
    plt.style.use(ele)
    plt.scatter(x,y1,color ='green',label = 'Quadratic',marker='^' )
    plt.scatter(x,y2,color ='red' ,label = 'Linear',marker='o')
    plt.xlabel('Time')
    plt.ylabel('Distance')
    plt.legend(loc='upper left')
    plt.title(ele)
    plt.show()
```

# **Bar Graph**

```
themes = plt.style.available
for ele in themes:
    plt.figure(figsize=(7,7))
    plt.style.use(ele)
    #plt.scatter(x,y1,color ='green',label = 'Quadratic',marker='^' )
    #plt.scatter(x,y2,color ='red' ,label = 'Linear',marker='o')
    plt.bar([1,2,3],[10,20,15],color ='green')
    plt.xlabel('Time')
    plt.ylabel('Distance')
    #plt.legend(loc='upper left')
    plt.title(ele)
    plt.show()
```

```
themes = plt.style.available
x_coordinates = np.array([0,1,2])*3
for ele in themes:
    plt.figure(figsize=(7,7))
    plt.style.use(ele)
    plt.bar(x_coordinates,[10,20,15],color ='green',width=1,label='current year', tick_label = ['Audi','Tata','Maruti'])
    plt.bar(x_coordinates+1,[20,5,15],color ='blue',width=1,label = 'next year')
    plt.xlabel('Time')
    plt.ylim(0,40)
    plt.xlim(-2,10)
    plt.ylabel('Distance')
    plt.legend(loc='best')
    plt.title(ele)
    plt.show()
```

# **Normal Distribution**
*  Mean
*  Standard Deviation

```
xsn = np.random.randn(100)
sigma = 5
u=80
x = np.round(xsn*sigma + u)
x2 = np.round(xsn*5 + 40)

print(x,x2)
```

# **Histogram**

```
plt.style.use("seaborn")
plt.hist(x,label='Math')
plt.hist(x2,label='Physics')
plt.xlabel('Marks Range')
plt.ylabel('Probability of students')
plt.title('Histogram')
plt.legend()
plt.show()

# use alpha = 0.8 for opacity in case of overlapping
```
